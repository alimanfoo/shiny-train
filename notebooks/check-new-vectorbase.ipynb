```python
import pyfasta
from pyprojroot import here
import allel
```

```python
genome_old = pyfasta.Fasta(str(here() / 'data/external/vectorbase/Anopheles-gambiae-PEST_CHROMOSOMES_AgamP4.fa'))
genome_old
```

```python
for k in sorted(genome_old):
    print(k, len(genome_old[k]))
```

```python
def key_fn(s):
    k = s.split()[0]
    if k.startswith('AgamP4'):
        k = k.split('_')[1]
    return k

genome_new = pyfasta.Fasta(str(),
                           key_fn=key_fn)
genome_new
```

```python
for k in sorted(genome_new):
    if not k.startswith('AAAB'):
        print(k, len(genome_new[k]))
```

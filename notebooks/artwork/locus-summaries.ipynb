```python
def load_cnv_df():

    df_cnv = (
        pd.read_excel(
            here() / 'data/external/lucas_2019/Supplementary_Data_S9.xls',
            skiprows=4,
            index_col=0
        )
        .rename(columns={
            'Angola\ncoluzzii\n(78)': 'ao_col',
            'Burkina Faso\ncoluzzii\n(75)': 'bf_col',
            'Burkina Faso\ngambiae\n(92)': 'bf_gam',
            "Côte d'Ivoire\ncoluzzii\n(71)": 'ci_col',
            "Cameroon\ngambiae\n(297)": 'cm_gam',
            "Mayotte\ngambiae\n(24)": 'fr_gam',
            "Gabon\ngambiae\n(69)": 'ga_gam',
            'Ghana\ncoluzzii\n(55)': 'gh_col', 
            'Ghana\ngambiae\n(12)': 'gh_gam',
            'The Gambia\nundetermined\n(65)': 'gm', 
            'Guinea\ncoluzzii\n(4)': 'gn_col',
            'Guinea\ngambiae\n(40)': 'gn_gam', 
            'Equ. Guinea\ngambiae\n(9)': 'gq_gam',
            'Guinea-Bissau\nundetermined\n(91)': 'gw', 
            'Kenya\nundetermined\n(48)': 'ke',
            'Uganda\ngambiae\n(112)': 'ug_gam',
        })
        .T
    )

    gstue_cols = [n for n in df_cnv.columns if n.startswith('Gstue')]
    cyp6aap_cols = [n for n in df_cnv.columns if n.startswith('Cyp6aap')]
    cyp9k1_cols = [n for n in df_cnv.columns if n.startswith('Cyp9k1')]
    df_cnv['cyp6aap_cnv_fmax'] = df_cnv[cyp6aap_cols].max(axis=1)
    df_cnv['cyp9k1_cnv_fmax'] = df_cnv[cyp9k1_cols].max(axis=1)
    df_cnv['gstue_cnv_fmax'] = df_cnv[gstue_cols].max(axis=1)
    df_cnv = df_cnv[['gstue_cnv_fmax', 'cyp6aap_cnv_fmax', 'cyp9k1_cnv_fmax']]
    return df_cnv

df_cnv = load_cnv_df()
df_cnv
```

```python
def load_gste2_i144t_df():

    gste2_i114t_pos = 28598166

    import intake
    cat = intake.open_catalog('https://malariagen.github.io/intake/gcs.yml')
    callset_snps = cat.ag2.snps.to_zarr()
    pos = allel.SortedIndex(callset_snps['3R/variants/POS'])
    gste2_i114t_loc = pos.locate_key(gste2_i114t_pos)
    gste2_i114t_gt = allel.GenotypeArray(callset_snps['3R/calldata/GT'][gste2_i114t_loc][None, :, :])
    df_samples = cat.ag2.samples.read()   
    pop_rename = {
        p: p.lower() if len(p) == 2 else p.lower()[:2] + '_' + p[2:]
        for p in df_samples.population.unique()
    }
    df_samples.population.replace(pop_rename, inplace=True)
    pop_locs = {
        p: df_samples[df_samples.population == p].index.values
        for p in df_samples.population.unique()
    }
    gste2_i114t_acs = gste2_i114t_gt.count_alleles_subpops(pop_locs)
    gste2_i114t_freqs = [
        {'pop': p, 'gste2_i114t_f': gste2_i114t_acs[p].to_frequencies()[0, 1]}
        for p in df_samples.population.unique()
    ]
    df = pd.DataFrame.from_records(gste2_i114t_freqs, index='pop')
    return df

df_gste2_i114t = load_gste2_i144t_df()
df_gste2_i114t
```
